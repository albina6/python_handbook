```python
## C

with open('C.txt', encoding='UTF-8') as file_in:
    lines = file_in.readlines()
prog = []
for i in lines:
    if '#' in i:
        index_ = i.find('''#''')
        if index_ > 0:
            if (c:=prog.count('\'\'\'')) > 0 and c % 2 != 0 or ((b:=i[:index_].count('\'\'\'')) > 0 and b % 2 != 0) :
                prog.append(i)
                continue
            if '\'' in i:
                if i[:index_].count('\'') % 2 != 0:
                    prog.append(i)
                    continue
            if '\"' in i:
                if i[:index_].count('\"') % 2 != 0:
                    prog.append(i)
                    continue
            
            prog.append(i[:index_])
    else:
        prog.append(i.rstrip("\n"))
for i in prog:
    print(i)
```

```python
## L

def rec_file(name_file, lines):
    lines.append('')
    lines_ = '\n'.join(lines)
    with open(name_file, "w", encoding='UTF-8') as file_out:
        file_out.write(lines_)  

file_in, file_even, file_odd, file_eq = input(), input(), input(), input()

# file_in, file_even, file_odd, file_eq = 'L.txt', 'file_even.txt', 'file_odd.txt' , 'file_eq.txt'

with open(file_in, encoding='UTF-8') as file_in:
    lines = file_in.readlines()

even_list = []
odd_list = []
eq_list = []
for i in lines:
    k = (i.rstrip("\n").split())

    even_ = []
    odd_ = []
    eq_ = []
    
    for j in k:
        even = sum(1 for n in j if int(n) % 2 == 0)
        odd = len(j) - even
        if even > odd:
            even_.append(j)
        elif even < odd:
            odd_.append(j)
        else:
            eq_.append(j)
    even_list.append(' '.join(even_))
    odd_list.append(' '.join(odd_))
    eq_list.append(' '.join(eq_))

rec_file(file_even, even_list)
rec_file(file_odd, odd_list)
rec_file(file_eq, eq_list)
```

```python
## N

import json


def read_json(file_name):
    with open(file_name, encoding='UTF-8') as file_in:
        lines = json.load(file_in)
    return lines


file_users = input()
file_data = input()

# file_users = 'N_users.json'
# file_data = 'N_updates.json'

user_f = read_json(file_users)
data_f = read_json(file_data)

for d in data_f:
    name = d['name']
    for user in user_f:
        if name == user['name']:
            for key, value in d.items():
                val_new = max(user.get(key, value), value)
                user[key] = val_new
                
dict_ = {}
for i in user_f:
    dict_[i['name']] = {}
    for key, value in i.items():
        if key != 'name':
            dict_[i['name']][key] = value

with open(file_users, 'w', encoding='UTF-8') as file_out:
    json.dump(dict_, file_out, ensure_ascii=False, indent=2)
```

# P


```python
## P

from sys import stdin


def read_file(name_file):
    with open(name_file, encoding='UTF-8') as file_in:
        lines = file_in.readlines()
    return lines


def clear_(text):
    new_text = []
    k = ' '
    space_ = [' ', '\n', '\t']
    for line in text:
        for letter in line:
            if letter in space_:
                if k == ' ':
                    continue
                else:
                    letter = ' '
            new_text += letter.lower()
            k = letter
    return ''.join(new_text)


def find_req(request, lines):
    return clear_(request) in clear_(lines)
    

request = input()
name_files = []
for line in stdin:
    name_files.append(line.rstrip('\n'))

# request = 'Мама мыла РАМУ'
# name_files = ['P_first.txt', 'P_second.txt']
flag = True
for i in name_files:
    lines = read_file(i.rstrip('\n'))
    if find_req(request, lines):
        flag = False
        print(i)

if flag:
    print('404. Not Found')
```

```python
## S

import string


lat_letters = string.ascii_lowercase

shift_n = int(input())

with open('public.txt', encoding='UTF-8') as file_in:
    lines = file_in.readlines()

# shift_n = 26
# lines ='Hello, world!'
text = []
for line in lines:
    for letter in line:
        if letter.lower() in lat_letters:
            k = (lat_letters.index(letter.lower()) + shift_n) % 26
            let_ = lat_letters[k]
            text += let_ if letter.islower() else let_.upper()
        else:
            text += letter
            
# print(*text, sep='')

text = ''.join(text)

with open('private.txt', 'w', encoding='UTF-8') as file_out:
    file_out.writelines(text)
```
